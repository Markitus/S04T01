```python
import pandas as pd
import matplotlib.pyplot as plt

coses = pd.read_csv('DelayedFlights.csv')
fig1=coses['UniqueCarrier'].value_counts().plot(kind="bar",title = 'Vols/Aerolinia')

plt.show()
```

```python
fig1.get_figure().savefig('fig1.png')
```

```python
fig2=coses['ArrDelay'].value_counts().plot(kind="hist", title = 'Vols/Pilots vagos')

plt.show()
```

```python
fig2.get_figure().savefig('fig2.png')
```

```python
import seaborn as sns

fig3=sns.barplot(y="ArrDelay", x="UniqueCarrier", data=coses)

plt.show()
```

```python
fig3.get_figure().savefig('fig3.png')
```
